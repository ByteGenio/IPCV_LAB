```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Read image in grayscale
img = cv2.imread('IMG_20250704_102853842 (1).jpg', cv2.IMREAD_GRAYSCALE)

# Check if image loaded
print(img.shape)

plt.imshow(img, cmap='gray')
plt.title("Original Image")
plt.axis('off')
```

```python
negative = 255 - img

plt.imshow(negative, cmap='gray')
plt.title("Image Negative")
plt.axis('off')
```

```python
r_min = np.min(img)
r_max = np.max(img)

contrast = ((img - r_min) / (r_max - r_min)) * 255
contrast = contrast.astype(np.uint8)

plt.imshow(contrast, cmap='gray')
plt.title("Contrast Stretched Image")
plt.axis('off')
```

```python
bit_planes = []

for i in range(8):
    bit = (img >> i) & 1
    bit_planes.append(bit * 255)

plt.figure(figsize=(12,6))
for i in range(8):
    plt.subplot(2,4,i+1)
    plt.imshow(bit_planes[i], cmap='gray')
    plt.title(f'Bit Plane {i}')
    plt.axis('off')
plt.show()
```

```python
lower = 100
upper = 150

gray_slice = np.zeros_like(img)
gray_slice[(img >= lower) & (img <= upper)] = 255

plt.imshow(gray_slice, cmap='gray')
plt.title("Gray Level Slicing")
plt.axis('off')
```
